```python
DataBDS = DataBDS.dropna(subset=['Gi√°'])
DataBDS['Pricem2'] = DataBDS[DataBDS['Gi√°'].str.contains('tri·ªáu', case=False)]['Gi√°'].str.extract('(\d+)').astype(float)
DataBDS = DataBDS[DataBDS['Pricem2'] >= 30]
DataBDS['Location']=DataBDS['V·ªã Tr√≠'].str.split(',').str[-2].str.strip()
DataBDS['Type'] = DataBDS['Lo·∫°i'].apply(lambda x: 0 if pd.notna(x) and 'chung c∆∞' in x else 1)
DataBDS['Acreage'] = DataBDS['Di·ªán t√≠ch'].str.extract('(\d+)').astype(float)
DataBDS['Direction'] = DataBDS['H∆∞·ªõng'].apply(lambda x: 1 if pd.notna(x) and 'ƒê√¥ng' in x else 0)
DataBDS['BathRoom']=DataBDS['Tolet'].fillna(0)
DataBDS['FrontHouse'] = (DataBDS['ƒê∆∞·ªùng'].str.extract('(\d+)').astype(float)).fillna(0)
DataBDS['LivingRoom']=DataBDS['Ph√≤ng kh√°ch'].fillna(0)
DataBDS['BedRoom'] = (DataBDS['Ph√≤ng ng·ªß'].str.extract('(\d+)').astype(float)).fillna(0)
DataBDS['Juridical'] = (DataBDS['Ph√°p l√Ω']).apply(lambda x: 1 if not pd.isna(x) else 0)
DataBDS.dropna(subset=['Pricem2'], inplace=True)
DataBDS['Price'] = DataBDS['Pricem2'] * DataBDS['Acreage']
```

```python
DataBDS = DataBDS.drop(columns=['Unnamed: 0','Title', 'Th·ªùi gian', 'V·ªã Tr√≠', 'Gi√°', 'Di·ªán t√≠ch', 'Lo·∫°i','H∆∞·ªõng','t·∫ßng','Tolet','ƒê∆∞·ªùng','Ph√≤ng kh√°ch','Ph√≤ng ng·ªß','Ph√°p l√Ω','N·ªôi dung','Pricem2'])
```

```python
input_cols = ['Type', 'Acreage',  'Direction', 'BathRoom',
       'FrontHouse', 'LivingRoom', 'BedRoom','Juridical']
target = ['Price']
inputs_df = DataBDS[input_cols]
target_df = DataBDS[target]
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, LambdaCallback
X_train, X_test, y_train, y_test = train_test_split(inputs_df, target_df, test_size = 0.1, shuffle = True, random_state=42)
```

```python
print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
from keras.models import load_model
def get_model():
    model = Sequential([
        Dense(8, input_shape = (8,), activation = 'relu'),
        Dense(16, activation = 'relu'),
        Dense(4, activation = 'relu'),
        Dense(1)
    ])

    model.compile(
        loss='mse',
        optimizer='adam'
    )

    return model

model = get_model()
model.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience = 10)

model = get_model()

history = model.fit(
    X_train, y_train,
    validation_data = (X_test, y_test),
    epochs = 100,
    callbacks = [early_stopping]
)
preds = model.predict(X_test)
model.save("modelpredictBDS.h5")
```
